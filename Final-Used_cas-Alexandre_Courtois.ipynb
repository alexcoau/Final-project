## Prep for machine learning


```python
ds_type = ds[['vehicleType']]
ds_gb = ds[['gearbox']]
ds_model = ds[['model']]
ds_fuel = ds[['fuelType']]
ds_brand = ds[['brand']]
```

### Split

```python
import numpy as np

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]

train_set, test_set = split_train_test(ds, 0.2)
print(len(train_set), "train +", len(test_set), "test")

ds = train_set.drop("price", axis=1) # drop labels for training set
ds_labels = train_set["price"].copy()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy="median")

# We only keep numerical values
ds_num = ds.select_dtypes(include=[np.number])
imputer.fit(ds_num)
```

```python
ds_num = ds_num[~ds_num.index.duplicated(keep='first')]
X = imputer.transform(ds_num)
ds_tr = pd.DataFrame(X, columns=ds_num.columns,
                          index = list(ds.index.values))
ds_tr.tail()
```

## Models

#### Linear regressor

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(ds_num, ds_labels)
```

```python
some_data = ds_num.iloc[:5]
some_labels = ds_labels.iloc[:5]
#some_data_prepared = full_pipeline.transform(some_data)

print("Predictions:", lin_reg.predict(some_data))
print("labels:", some_labels)
```

```python
from sklearn.metrics import mean_squared_error,  mean_absolute_error

ds_predictions = lin_reg.predict(ds_num)
lin_mse = mean_squared_error(ds_labels, ds_predictions)
lin_rmse = np.sqrt(lin_mse)
print("RMSE : ",lin_rmse)
lin_mae = mean_absolute_error(ds_labels, ds_predictions)
print("MAE : ",lin_mae)
```

#### Decision tree

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(random_state=29)
tree_reg.fit(ds_num, ds_labels)
```

```python
some_data = ds_num.iloc[:5]
some_labels = ds_labels.iloc[:5]
#some_data_prepared = full_pipeline.transform(some_data)

print("Predictions:", tree_reg.predict(some_data))
print("labels:", some_labels)
```

```python
ds_predictions_tree = tree_reg.predict(ds_num)
lin_mse = mean_squared_error(ds_labels, ds_predictions_tree)
lin_rmse = np.sqrt(lin_mse)
print("RMSE : ",lin_rmse)
lin_mae = mean_absolute_error(ds_labels, ds_predictions_tree)
print("MAE : ",lin_mae)
```

#### Random forest

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(random_state=29)
forest_reg.fit(ds_num, ds_labels)
```

```python
some_data = ds_num.iloc[:5]
some_labels = ds_labels.iloc[:5]
#some_data_prepared = full_pipeline.transform(some_data)

print("Predictions:", forest_reg.predict(some_data))
print("labels:", some_labels)
```
